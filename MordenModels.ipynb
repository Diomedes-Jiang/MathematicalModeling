```python
result=[]
result.append(dis)
selfChange=[]
selfChange.append(100)
def annealing(dis,distance,ans):
    Temperature=100
    T=1e-20
    alpha=0.99
    while T<Temperature:
        neoAns=np.asarray(ans)
        flag=random.randint(0,1)
        if flag==0: # 2变换
            first=second=1
            first=random.randint(0,len(distance)-1)
            while first==second:
                second=random.randint(0,len(distance)-1)
            subAns=neoAns[first:second]
            subAns=subAns[::-1]
            neoAns[first:second]=subAns
        if flag==1: # 3变换
            one=two=three=0
            while one>=two or two>=three or three==len(distance)-1:
                one=random.randint(0,len(distance)-1)
                two=random.randint(one,len(distance)-1)
                three=random.randint(two,len(distance)-1)
            subAns=list(neoAns[one:two])
            length=two-one
            for i in range(two,three+1):
                neoAns[i-length]=neoAns[i]
            neoAns[three+1-length:three+1]=subAns
        
        neoDis=calDis(neoAns,distance)
        if neoDis<dis:# 如果是最优解，则接受
            dis=neoDis
            ans=neoAns
        else: # 如果不是最优解，则以概率接受
            print(f"neoDis:{neoDis},dis:{dis}")
            Possibility=np.exp(-(neoDis-dis)/Temperature)
            P=np.random.rand()
            print(f"Possibility: {Possibility}, P: {P}")
            if P < Possibility: # 接受
                dis=neoDis
                ans=neoAns
        Temperature=alpha*Temperature 
        selfChange.append(Temperature)
        result.append(dis)
        print(f"when the temperature is {Temperature}, the minimum distance is {dis}")
```

# 遗传算法

```python
population=50
Generation=1000
crossoverRate=1
mutationRate=0.1
```

```python
array=np.arange(1,101)
np.random.shuffle(array[1:99])
array
```

```python
def initialize(population):
    Group=np.zeros((population,100))
    for i in range(population):
        temp=np.arange(1,101)
        np.random.shuffle(temp[1:99])
        temp_distance=calDis(temp,distance)
        while True:
            u=v=1
            while u>=v :
                u=random.randint(1,98)
                v=random.randint(u,98)
            neoTemp=list(temp)
            subTemp=neoTemp[u:v]
            subTemp=subTemp[::-1]
            neoTemp[u:v]=subTemp
            if calDis(neoTemp,distance)<temp_distance:
                Group[i]=neoTemp
                break
    return np.int64(Group)
```

```python
def encoding(Group):
    chromosome=np.random.rand(50,100)
    for i in range(50):
        chromosome[i]=np.sort(chromosome[i])
        sorted_indices=Group[i]-1
        chromosome[i,]=chromosome[i,sorted_indices]
    return chromosome
```

```python
def greater(array):
    group=np.zeros(len(array))
    for i in range(len(array)):
        posi=np.arange(1,101)
        pos=0
        for j in range(len(array)):
            if array[i]>array[j]:
                pos+=1
        while posi[pos]==0:
            pos+=1
        group[i]=posi[pos]-1
        posi[pos]=0
    return np.int64(group)    

def calFitness(chromosomes):
    fitness=np.zeros(chromosomes.shape[0])
    for i in range(chromosomes.shape[0]):
        group=greater(chromosomes[i])
        f=0
        for j in range(len(group)-1):
            f+=distance[group[j],group[j+1]]
        fitness[i]=f
    return fitness
```

```python
def crossover(chromosome1,chromosome2):
    t=random.randint(1,len(chromosome1)-1)
    parent1=chromosome1[:t]
    parent2=chromosome2[:t]
    child1=chromosome1[t:]
    child2=chromosome2[t:]
    neo_chromosome1=np.concatenate((parent1,child2))
    neo_chromosome2=np.concatenate((parent2,child1))
    while len(np.unique(neo_chromosome1))<len(neo_chromosome1):
        unique_elements, first_occurrence_indices = np.unique(neo_chromosome1, return_index=True)
        # 找到重复元素的下标
        duplicate_indices = np.setdiff1d(np.arange(len(neo_chromosome1)), first_occurrence_indices)
        for i in range(len(duplicate_indices)):
            neo_chromosome1[duplicate_indices[i]]=random.random()
            
    while len(np.unique(neo_chromosome2))<len(neo_chromosome2):
        unique_elements, first_occurrence_indices = np.unique(neo_chromosome2, return_index=True)
        # 找到重复元素的下标
        duplicate_indices = np.setdiff1d(np.arange(len(neo_chromosome2)), first_occurrence_indices)
        for i in range(len(duplicate_indices)):
            neo_chromosome2[duplicate_indices[i]]=random.random()
    return neo_chromosome1,neo_chromosome2
```

```python
def mutation(chromosome):
    r=random.random()
    if(r<mutationRate):
        one=two=three=0
        while one>=two or two>=three or three==len(chromosome)-1:
            one=random.randint(0,len(chromosome)-1)
            two=random.randint(one,len(chromosome)-1)
            three=random.randint(two,len(chromosome)-1)
        subChr=list(chromosome[one:two])
        length=two-one
        for i in range(two,three+1):
            chromosome[i-length]=chromosome[i]
        chromosome[three+1-length:three+1]=subChr
    # return chromosome
    
```

```python
def selection(chromosomes):
    fitness=calFitness(chromosomes)
    order=np.argsort(fitness)
    order=order[:50]
    neo_chromosomes=np.zeros((50,100))
    for i in range(50):
        neo_chromosomes[i]=chromosomes[order[i]]
    return neo_chromosomes
        
```

```python
def GeneticAlgorithem():
    # 在TSP问题上性能接近scikit包
    result=[]
    cricles=np.arange(0,1000)
    Group=initialize(population)
    Group=np.int64(Group)
    chromosomes=encoding(Group)
    circle=0
    while circle<Generation:
        for i in range(50):
            for j in range(i,50):
                chr1,chr2=crossover(chromosomes[i],chromosomes[j])
                chromosomes=np.vstack((chromosomes,chr1[np.newaxis,:]))
                chromosomes=np.vstack((chromosomes,chr2[np.newaxis,:]))
        for i in range(50,len(chromosomes)):
            mutation(chromosomes[i])
        chromosomes=selection(chromosomes)
        group=greater(chromosomes[0])
        circle+=1
        print(f"when the circle is {circle} distance is {min(calFitness(chromosomes))}, the sequence is {group}")
        result.append(min(calFitness(chromosomes)))
    plt.plot(cricles,result)
    plt.show()
GeneticAlgorithem()
        
```
