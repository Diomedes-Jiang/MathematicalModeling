# PCA
PCA的主要目标是减小特征维度，主要方法为：  
- 换特征
- 减少特征

通过旋转坐标轴将高维空间的原始样本点投影到低维空间，减少维度特征，达到降维的目的

```python
import numpy as np
# 构造数据
X=np.random.randint(10,2000,size=[18,6])
X
```

```python
from scipy import linalg
```

```python
X_standardized=(X - np.mean(X,axis=0))/np.std(X,ddof=1,axis=0)
# ddof即为计算总体或样本标准差，ddof=0表示计算总体标准差 最终除以n，ddof=1计算样本标准差，除以n-1
X_standardized
```

```python
R=np.cov(X_standardized.T)
R
```

```python
#计算特征值特征向量
val,vec=linalg.eigh(R) # 该函数默认给出升序
print(val)
print(vec)
val=val[::-1]
vec=vec[:,::-1]
print(val)
print(vec)
```

```python
contribute_value=val/sum(val) # 计算出每个特征值对应的贡献率
cum_contribute_value=np.cumsum(contribute_value) # 计算出累计贡献率
contribute_value,cum_contribute_value
```

可以看到，在第四个累计贡献率处达到了80%以上
则可以将最初六个相关性转换为上述对应四个特征向量的线性组合
