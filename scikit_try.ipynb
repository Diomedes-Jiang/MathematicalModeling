```python
def cal_total_distance(points):
    num_points, = points.shape
    total_distance = 0
    for i in range(num_points - 1):
        total_distance += distance[points[i], points[i + 1]]
    total_distance += distance[points[i + 1], points[0]]
    return total_distance
```
